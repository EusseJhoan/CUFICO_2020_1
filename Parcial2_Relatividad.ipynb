**Variedad**

Usaremos la superficie:
$$E=\{ (x,y,z) \in \mathbb{R}^3 | \, \frac{x^2}{a^2}+\frac{y^2}{a^2}+\frac{z^2}{b^2} =1 \}$$

Definimos las siguientes conjuntos abiertos para construir las cartas que forman nuestra variedad:

$$U_{x_{\lessgtr 0}}= \{ (x,y,z) \in E: x\lessgtr0 \} $$

$$U_{y_{\lessgtr 0}}= \{ (x,y,z) \in E: y\lessgtr0 \} $$

$$U_{z_{\lessgtr 0}}= \{ (x,y,z) \in E: z\lessgtr0 \} $$

con los respectivos mapas:

$$\varphi_{x_{\lessgtr 0}} : U_{x_{\lessgtr 0}} \rightarrow \mathbb{R}^2 \, | \, \varphi_{x_{\lessgtr 0}(x,y,z)} = (y,z)$$ 

$$\varphi_{y_{\lessgtr 0}} : U_{y_{\lessgtr 0}} \rightarrow \mathbb{R}^2 \, | \, \varphi_{y_{\lessgtr 0}(x,y,z)} = (x,z)$$ 

$$\varphi_{z_{\lessgtr 0}} : U_{z_{\lessgtr 0}} \rightarrow \mathbb{R}^2 \, | \, \varphi_{z_{\lessgtr 0}(x,y,z)} = (x,y)$$ 

Las 6 cartas foman así el atlas:

$$U=\{(U_{x_{> 0}},\varphi_{x_{> 0}}), (U_{y_{> 0}},\varphi_{y_{> 0}}),(U_{z_{> 0}},\varphi_{z_{> 0}}),(U_{x_{< 0}},\varphi_{x_{< 0}}),(U_{y_{< 0}},\varphi_{y_{< 0}}),(U_{z_{< 0}},\varphi_{z_{< 0}}) \}$$


Veamos que este atlas, efectivamente, da lugar a un espacio topológico: 

Asumimos que $E$ está sumergido en un espacio métrico de $\mathbb{R}^3$, es fácil demostrar que $\mathbb{R}^3$ cumple todas las propiedades de espacio métrico. Es claro que $\cup_{i}U_{i} = E$, es decir, $\{U_{i}\}$ es una familia de abiertos que cubren $E$. Entonces el conjunto

$$\tau = \{U_{i} \subset E: \forall \hspace{0.1cm}(x,y,z) \in U_{i}\hspace{0.1cm} \exists\hspace{0.1cm} \varepsilon > 0 \hspace{0.1cm} | \, \hspace{0.1cm} B_{\varepsilon}(x,y,z) \subset U_{i} \}$$


define una topología de $E$, donde $B_{\varepsilon}(x,y,z)$ es una bola abierta de radio $\varepsilon$ centrada en el punto $(x,y,z)$. Luego $\tau$ cumple todas las propiedades de espacio topológico.

Dados los mapas $\varphi$, podemos definir los mapas inversos:


$$\varphi_{x_{\lessgtr 0}}^{-1} :(y,z) \rightarrow \left(\mp a \sqrt{1-(\frac{z^2}{b^2}+\frac{y^2}{a^2})},y,z \right)$$ 

$$\varphi_{y_{\lessgtr 0}}^{-1} :(x,z)\rightarrow \left(x, \mp a \sqrt{1-(\frac{x^2}{a^2}+\frac{z^2}{b^2})},z \right)$$ 


$$\varphi_{z_{\lessgtr 0}}^{-1} :(x,y) \rightarrow \left(x,y,\mp b \sqrt{1-(\frac{x^2}{a^2}+\frac{y^2}{a^2})}\right)$$ 

Vemos que las funciones $\varphi_{x_{\lessgtr 0}}$, $\varphi_{y_{\lessgtr 0}}$ y $\varphi_{z_{\lessgtr 0}}$ son continuas, biyectivas y que además sus inversas $\varphi_{x_{\lessgtr 0}}^{-1}$, $\varphi_{y_{\lessgtr 0}}^{-1}$ y $\varphi_{z_{\lessgtr 0}}^{-1}$ son también continuas para los dominios dados por la variedad en la que trabajamos por tanto las funciones $\varphi$  son un homeomorfismo.

Además, tomemos como ejemplo dos cartas tales que la intersecccion de sus conjuntos abiertos $U$ asociados es diferente del vacío por ejemplo $\{U_{x>0},\varphi_{x>0}\}$ y $\{U_{y>0},\varphi_{y>0}\}$

y definamos $\psi_{_{x>0\,y>0}}= \varphi_{_{x>0}}\circ \varphi_{_{y>0}}^{-1}$ así:

$$\psi_{_{x>0\,y>0}}(x,z)= \varphi_{_{x>0}}\circ \varphi_{_{y>0}}^{-1}(x,z) $$
$$\psi_{_{x>0\,y>0}}(x,z)= \varphi_{_{x>0}}[ \varphi_{_{y>0}}^{-1}(x,z)] = \varphi_{_{x>0}}\left [\left(x,a \sqrt{1-(\frac{x^2}{a^2}+\frac{z^2}{b^2})},z \right)\right]$$
$$\psi_{_{x>0\,y>0}}(x,z)=\left(a \sqrt{1-(\frac{x^2}{a^2}+\frac{z^2}{b^2})},z \right)$$

De manera similar: 
$$\psi_{_{y>0\,x>0}}(y,z)=\varphi_{_{y>0}}\circ \varphi_{_{x>0}}^{-1}(y,z)= \left(a \sqrt{1-(\frac{y^2}{a^2}+\frac{z^2}{b^2})},z \right)$$

Que son infinitamente diferenciables.Se puede mostrar de manera similar que para cualquier par de cartas del atlas $U$ que cumplan la condición $U_i\cap U_j \neq \phi$  la respectiva función $\psi_{ij}$ que va de una a otra es infinitamente diferenciable.

De todos los argumentos anteriormente expuestos podemos afimrar que la variedad estudiada $E$ es una variedad diferenciable que es convexa y compacta.

**Relación_R3**

Si nuestra variedad viviera sumergida en un espacio $\mathbb{R}^3$, podemos usar sistema de coordenadas adecuado que describa la variedad. Si usamos coordenadas esféricas $(r,\theta,\phi)$, tenemos la superficie

$$E^{\prime}=\{ (r,\theta,\phi) \in (0,\infty)\times(0,\pi)\times[0,2\pi) | \, r =\dfrac{c}{\sqrt{1 - e^{2}\sin^{2}\theta}} \}$$
donde $e^{2} = (a^{2} - c^{2})/a^2$

Luego para $E^{\prime}$ podemos construir un atlas, que consta de una sola carta $\{V,\psi\}$, tal que

$$V= \{ (r,\theta,\phi) \in E^{\prime}: 0< \theta <\pi, \hspace{0.1cm} 0<\phi< 2\pi \} $$
\
$$\psi: V\rightarrow \mathbb{R}^2 \, | \, \psi(r,\theta,\phi) = (\theta,\phi)$$
\
$$\psi^{-1} :(\theta,\phi) \rightarrow \left(\dfrac{c}{\sqrt{1 - e^{2}\sin^{2}\theta}},\theta,\phi \right)$$

Luego, existe una correspondencia $f: E^{\prime} \rightarrow E$. De coordenadas esféricas a coordenadas cartesianas conocemos $f(r,\theta,\phi) = (x,y,z)$ como
 

$$x = r\sin\theta \cos\phi $$
$$y = r\sin\theta \sin\phi  $$
$$z = r\cos\theta  $$

Ahora tomemos, por ejemplo, las cartas $\{U_{x\lessgtr0},\varphi_{x\lessgtr0}\}$ de $E$. Luego la relación con el sistema de coordenadas esféricas está dada por el mapa $\varphi_{x_{\lessgtr 0}}\circ f\circ \psi^{-1}$, así


 \begin{array}{c} \varphi_{x_{\lessgtr 0}}(f(\dfrac{c}{\sqrt{1 - e^{2}\sin^{2}\theta}},\theta,\phi)) = (y,z) \\ \varphi_{x_{\lessgtr 0}}(\dfrac{c}{\sqrt{1 - e^{2}\sin^{2}\theta}}\sin\theta \cos\phi,\dfrac{c}{\sqrt{1 - e^{2}\sin^{2}\theta}}\sin\theta \sin\phi,\dfrac{c}{\sqrt{1 - e^{2}\sin^{2}\theta}}\cos\theta) = (y,z)\\\implies y = \dfrac{c}{\sqrt{1 - e^{2}\sin^{2}\theta}}\sin\theta \sin\phi \\ z = \dfrac{c}{\sqrt{1 - e^{2}\sin^{2}\theta}}\cos\theta  \end{array}

Por último para otras cartas se encuentra la relación para $x$, como
$$x = \dfrac{c}{\sqrt{1 - e^{2}\sin^{2}\theta}}\sin\theta \cos\phi$$

Definimos los campos vectoriales

$$\mathcal{V}=V_0\cos{\theta}\, \overrightarrow{e_{\phi}}$$

$$\mathcal{W}=W_0\sin{\phi}\, \overrightarrow{e_{\theta}}$$

Recordemos que las coordenadas para un elipsoide de revolución pueden ser escritas de manera paramétrica como:

$$x = a\sin\theta \cos\phi $$
$$y = a\sin\theta \sin\phi  $$
$$z = c\cos\theta  $$

si tomamos la carta $\{U_{x>0},\phi_{x>0}\}$ hallamos las componentes de $\mathcal{V}$ en esta como:

Para $\mathcal{V}^{y}$:
$$\mathcal{V}^{y}=\frac{\partial y}{\partial \phi}\mathcal{V}^{\phi}+ \frac{\partial y}{\partial \theta}\mathcal{V}^{\theta} $$

$$\mathcal{V}^{y}=\frac{\partial (a\sin{\theta}\sin{\phi})}{\partial \phi}(V_0\cos{\theta})+ \frac{\partial a\sin{\theta}\sin{\phi}}{\partial \theta}(0) $$

$$\mathcal{V}^{y}=V_0a\sin{\theta}\cos{\theta}\cos{\phi}$$  

Para $\mathcal{V}^{z}$:
$$\mathcal{V}^{z}=\frac{\partial z}{\partial \phi}\mathcal{}^{\phi}+ \frac{\partial z}{\partial \theta}\mathcal{V}^{\theta} $$

$$\mathcal{V}^{z}=\frac{\partial (b\cos{\theta})}{\partial \phi}V_0\cos{\theta}+ \frac{\partial (b\cos{\theta})}{\partial \theta}(0) $$

$$\mathcal{V}^{z}=0$$

si usamos las transformaciones cartesianas para las funciones trigonométricas:

$$\boxed{\mathcal{V}^{y}=V_0a\frac{xz}{(x^2+y^2+z^2)}}$$

$$\boxed{\mathcal{V}^{z}=0}$$

Ahora hallamos las componentes de $\mathcal{W}$:

Para $\mathcal{W}^{y}$:
$$\mathcal{W}^{y}=\frac{\partial y}{\partial \phi}\mathcal{W}^{\phi}+ \frac{\partial y}{\partial \theta}\mathcal{W}^{\theta} $$

$$\mathcal{W}^{y}=\frac{\partial (a\sin{\theta}\cos{\phi})}{\partial \phi}(0)+ \frac{\partial (a\sin{\theta}\cos{\phi})}{\partial \theta}(W_0\sin{\phi}) $$

$$\mathcal{W}^{y}=W_0a\cos{\theta}\cos{\phi}\sin{\phi}$$

para  $\mathcal{W}^{z}$:

$$\mathcal{W}^{z}=\frac{\partial z}{\partial \phi}\mathcal{W}^{\phi}+ \frac{\partial z}{\partial \theta}\mathcal{W}^{\theta} $$

$$\mathcal{W}^{y}=\frac{\partial (b\cos{\theta})}{\partial \phi}(0)+ \frac{\partial (b\cos{\theta})}{\partial \theta}(W_0\sin{\phi}) $$

$$\mathcal{W}^{z}=-W_0b\sin{\theta}\sin{\phi}$$

si usamos las transformaciones cartesianas para las funciones trigonométricas:

$$\boxed{\mathcal{W}^{y}=W_0a\frac{y^2z}{(x^2+y^2)\sqrt{x^2+y^2+z^2}}}$$

$$\boxed{\mathcal{W}^{z}=-W_0b\frac{y}{\sqrt{x^2+y^2+z^2}}}$$

De manera similar podemos hallar las compontentes de $\mathcal{V}$ y $\mathcal{W}$ en la carta $\{U_{z>0},\varphi_{z>0}\}$:

Para $\mathcal{V}^{x}$:
$$\mathcal{V}^{x}=\frac{\partial x}{\partial \phi}\mathcal{V}^{\phi}+ \frac{\partial x}{\partial \theta}\mathcal{V}^{\theta} $$

$$\boxed{\mathcal{V}^{x}=-V_0a\frac{yz}{(x^2+y^2+z^2)}}$$

Para $\mathcal{V}^{y}$:
$$\mathcal{V}^{y}=\frac{\partial y}{\partial \phi}\mathcal{V}^{\phi}+ \frac{\partial y}{\partial \theta}\mathcal{V}^{\theta} $$

$$\boxed{\mathcal{V}^{y}=V_0a\frac{xz}{(x^2+y^2+z^2)}}$$

Para $\mathcal{W}^{x}$:
$$\mathcal{W}^{x}=\frac{\partial x}{\partial \phi}\mathcal{W}^{\varphi}+ \frac{\partial x}{\partial \theta}\mathcal{W}^{\theta} $$

$$\boxed{\mathcal{W}^{x}=W_0a\frac{xyz}{(x^2+y^2)\sqrt{x^2+y^2+z^2}}}$$

$$\boxed{\mathcal{W}^{y}=W_0a\frac{y^2z}{(x^2+y^2)\sqrt{x^2+y^2+z^2}}}$$

Ahora calculemos la derivadas de Lee usando la carta $\{U_{x>0},\varphi_{x>0}\}$: 

primero $\mathcal{L}_{\mathcal{V}}\mathcal{W}$ que por componentes:

$$\mathcal{L}_{\mathcal{V}}\mathcal{W}^y=\mathcal{V}^{y}\partial_{y} \mathcal{W}^{y}-\mathcal{W}^{y}\partial_{y} \mathcal{V}^{y}+\mathcal{V}^{z}\partial_{z} \mathcal{W}^{y}-\mathcal{W}^{z}\partial_{z} \mathcal{V}^{y}$$

$$\mathcal{L}_{\mathcal{V}}\mathcal{W}^y=\mathcal{V}^{y}\partial_{y} (W_0a\frac{y^2z}{(x^2+y^2)\sqrt{x^2+y^2+z^2}})-\mathcal{W}^{y}\partial_{y} (V_0a\frac{xz}{(x^2+y^2+z^2)})+\mathcal{V}^{z}\partial_{z} \mathcal(W_0a\frac{y^2z}{(x^2+y^2)\sqrt{x^2+y^2+z^2}})-\mathcal{W}^{z}\partial_{z} (V_0a\frac{xz}{(x^2+y^2+z^2)})$$

$$\mathcal{L}_{\mathcal{V}}\mathcal{W}^y=\mathcal{V}^{y}\frac{a W_0 y z (2 x^4 - y^4 + y^2 z^2 + x^2 (y^2 + 2 z^2))}{(x^2 + y^2)^{3/2} (x^2 + y^2 + z^2)^2}-\mathcal{W}^{y}\frac{-2 a V_0 x y z}{(x^2 + y^2 + z^2)^2}+\mathcal{V}^{z}\frac{a W_0 y^2 (x^2 + y^2 - z^2)}{\sqrt{x^2 + y^2} (x^2 + y^2 + z^2)^2}-\mathcal{W}^{z}\frac{a V_0 x (x^2 + y^2 - z^2)}{(x^2 + y^2 + z^2)^2}$$

$$\boxed{\mathcal{L}_{\mathcal{V}}\mathcal{W}^y=a V_0 W_0 x y \left(  \frac{b (x^2 + y^2 - z^2)}{(x^2 + y^2 + z^2)^{5/2}}+\frac{a (2 x^2 + y^2) z^2}{(x^2 + y^2)^{3/2} (x^2 + y^2 + z^2)^2} \right )}$$

Ahora calculemos la otra componente:

$$\mathcal{L}_{\mathcal{V}}\mathcal{W}^z=\mathcal{V}^{y}\partial_{y} \mathcal{W}^{z}-\mathcal{W}^{y}\partial_{y} \mathcal{V}^{z}+\mathcal{V}^{z}\partial_{z} \mathcal{W}^{z}-\mathcal{W}^{z}\partial_{z} \mathcal{V}^{z}$$

$$\mathcal{L}_{\mathcal{V}}\mathcal{W}^z=\mathcal{V}^{y}\partial_{y} \mathcal{W}^{z}=$$

$$\mathcal{L}_{\mathcal{V}}\mathcal{W}^z=\mathcal{V^y}\left ( \frac{b W_0 y^2}{(x^2 + y^2 + z^2)^{3/2}}-\frac{b W_0}{\sqrt{x^2 + y^2 + z^2}}\right)$$

$$\mathcal{L}_{\mathcal{V}}\mathcal{W}^z=
\boxed{\frac{-a b V_0 W_0 x z (x^2 + z^2)}{(x^2 + y^2 + z^2)^{5/2}}}$$

Así: $$\mathcal{L}_{\mathcal{V}}\mathcal{W}=(\mathcal{L}_{\mathcal{V}}\mathcal{W}^y)e_y+(\mathcal{L}_{\mathcal{V}}\mathcal{W}^z)e_z $$

Ahora para la misma carta calculemos la detrivada de Lee $\mathcal{L}_{\mathcal{W}}\mathcal{V}$

$$\mathcal{L}_{\mathcal{W}}\mathcal{V}^y=\mathcal{W}^{y}\partial_{y} \mathcal{V}^{y}-\mathcal{V}^{y}\partial_{y} \mathcal{W}^{y}+\mathcal{W}^{z}\partial_{z} \mathcal{V}^{y}-\mathcal{V}^{z}\partial_{z} \mathcal{W}^{y}$$

$$\mathcal{L}_{\mathcal{W}}\mathcal{V}^y=\mathcal{W}^{y}\frac{-2 a V_0 x y z}{(x^2 + y^2 + z^2)^2}-\mathcal{V}^{y}\frac{a W_0 y z (2 x^4 - y^4 + y^2 z^2 + x^2 (y^2 + 2 z^2))}{(x^2 + y^2)^{3/2} (x^2 + y^2 + z^2)^2}+\mathcal{W}^{z}\frac{a V_0 x (x^2 + y^2 - z^2)}{(x^2 + y^2 + z^2)^2}-\mathcal{V}^{z}\frac{a W_0 y^2 (x^2 + y^2 - z^2)}{\sqrt{x^2 + y^2} (x^2 + y^2 + z^2)^2}$$

$$\boxed{\mathcal{L}_{\mathcal{W}}\mathcal{V}^y=-a V_0 W_0 x y \left(  \frac{b (x^2 + y^2 - z^2)}{(x^2 + y^2 + z^2)^{5/2}}+\frac{a (2 x^2 + y^2) z^2}{(x^2 + y^2)^{3/2} (x^2 + y^2 + z^2)^2} \right )}$$

$$\mathcal{L}_{\mathcal{W}}\mathcal{V}^z=-\mathcal{V}^{y}\partial_{y} \mathcal{W}^{z}=\frac{a b V_0 W_0 x z (x^2 + z^2)}{(x^2 + y^2 + z^2)^{5/2}}$$

Así: $$\mathcal{L}_{\mathcal{W}}\mathcal{V}=(\mathcal{L}_{\mathcal{W}}\mathcal{V}^y)e_y+(\mathcal{L}_{\mathcal{W}}\mathcal{V}^z)e_z $$

De donde comparando con las componentes halladas para $\mathcal{L}_{\mathcal{V}}\mathcal{W}$ vemos que:

$$\mathcal{L}_{\mathcal{W}}\mathcal{V}=(-\mathcal{L}_{\mathcal{V}}\mathcal{W}^y)e_y+(-\mathcal{L}_{\mathcal{V}}\mathcal{W}^z)e_z$$ 

por lo que claramente:
$$\boxed{\mathcal{L}_{\mathcal{W}}\mathcal{V}=-\mathcal{L}_{\mathcal{V}}\mathcal{W}}$$

Es decir, la derivada de Lee es antisimétrica en la carta $\{U_{x>0},\varphi_{x>0}\}$

$(\theta,\phi)$ tal que su dominio en $E$ está dado por la carta $\{U_{x>0},\varphi_{x>0}\}$ y es tal que $x = a\cos\theta \cos\phi$, $\hspace{0.1cm}y = a\cos\theta\sin\phi$,$\hspace{0.1cm}$ $z = c\sin\theta$. La medida se hacen desde el centro del elipsoide, tal que 
$(\theta,\phi) \in (-\dfrac{\pi}{2},\dfrac{\pi}{2})\times(0,2\pi)$.

$(\theta,\phi)$ tal que su dominio en $E$ está dado por la carta $\{U_{z>0},\varphi_{z>0}\}$ y es tal que $x = a\sin\theta \cos\phi$, $\hspace{0.1cm}y = a\sin\theta\sin\phi$,$\hspace{0.1cm}$ $z = c\cos\theta$. La medida se hacen desde el polo norte del elipsoide, tal que 
$(\theta ,\phi) \in (0,\pi)\times(0,2\pi)$.

Ahora calculamos las métricas inducidas en las cartas anteriores, así 

$$g_{\mu\nu}dx^{\mu}\otimes dx^{\nu} = \delta_{\alpha \beta}\dfrac{\partial f^{\alpha}}{\partial x^{\mu}}\dfrac{\partial f^{\beta}}{\partial x^{\nu}}dx^{\mu}\otimes dx^{\nu}$$
